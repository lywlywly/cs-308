```python
import csv
import matplotlib.pyplot as plt

x: list[int] = []
y: list[int] = []

with open("result.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        x.append(int(row[0]))
        y.append(int(row[1]))
```

```python
plt.scatter(x=x, y=y)
plt.title("Running time")
plt.xlabel("n")
plt.ylabel("milliseconds")
plt.savefig("runtime.png")
```

```python
plt.scatter(x=x, y=[c**0.5 for c in y])
plt.title("Running time")
plt.xlabel("n")
plt.ylabel("square root of milliseconds")
plt.savefig("squared_runtime.png")
```
